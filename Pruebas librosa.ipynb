```python
mfcc = []
for index, row in df.iterrows():
    path = f"data/{row['genre']}/{row['title']}"
    '''print(path)'''
    
    y, sr = librosa.load(path, duration=15)
    mfcc.append(librosa.feature.mfcc(y=y, sr=sr))
mfcc[0]
```

```python
df['MFCC'] = mfcc
df
```

```python
#Intento de funcion para features:
def new_feature(feat,dataframe=df):
    new_feature_data = []
    for index, row in dataframe.iterrows():
        path = f"data/{row['genre']}/{row['title']}"

        y, sr = librosa.load(path, duration=15)

        new_feature_data.append(librosa.feature.feat(y=y, sr=sr))
        
        df[f"{feat}"] = new_feature_data
```

```python
tempogram_data = []

for index, row in df.iterrows():
    path = f"data/{row['genre']}/{row['title']}"

    y, sr = librosa.load(path, duration=15)

    tempogram_data.append(librosa.feature.tempogram(y=y, sr=sr))
        
df['Tempogram'] = tempogram_data
    
df.head()
```

```python
poly_features_data = []

for index, row in df.iterrows():
    path = f"data/{row['genre']}/{row['title']}"

    y, sr = librosa.load(path, duration=15)

    poly_features_data.append(librosa.feature.poly_features(y=y, sr=sr))
        
df['Poly_features'] = poly_features_data
    
df.head()
```

```python
plp = []

for index, row in df.iterrows():
    path = f"data/{row['genre']}/{row['title']}"

    y, sr = librosa.load(path, duration=15)

    plp.append(librosa.beat.plp(y=y, sr=sr))
        
df['Plp'] = plp
    
df.head()
```

```python
spectral_flatness = []

for index, row in df.iterrows():
    path = f"data/{row['genre']}/{row['title']}"

    y, sr = librosa.load(path, duration=15)

    spectral_flatness.append(librosa.feature.spectral_flatness(y=y))
        
df['Spectral_flatness'] = spectral_flatness
    
df.head()
```
